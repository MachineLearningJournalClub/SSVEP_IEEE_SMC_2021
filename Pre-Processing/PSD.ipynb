```python
import mne
from mne.preprocessing import (ICA)
import matplotlib.pyplot as plt
import numpy as np

data = mne.io.read_raw_fif("D:\Desktop\SSVEP/new_data/filt_ica_car.fif", preload = True)
data.plot(start=1, duration=50)

# Construct epochs
event_id = {'9Hz': 1, '10Hz': 2, '12Hz': 3, '15Hz': 4}
events = mne.find_events(data, stim_channel='stim')
baseline = None

tmin= 0
tmax = 7.35  # in s

epochs = mne.Epochs(
        data, events=events,
        event_id=event_id, tmin=tmin,
        tmax=tmax, baseline=baseline, verbose=False)

data9 = epochs['9Hz']
data10 = epochs['10Hz']
data12 = epochs['12Hz']
data15 = epochs['15Hz']

#%% Plot PSD per epochs

data9.plot_psd(fmin = 3, fmax = 32)

data10.plot_psd(fmin = 3, fmax = 32)

data12.plot_psd(fmin = 3, fmax = 32)

data15.plot_psd(fmin = 3, fmax = 32)
```
