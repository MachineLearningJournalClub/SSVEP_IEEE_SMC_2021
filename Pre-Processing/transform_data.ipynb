```python
import numpy as np
np_raw = np.load('D:\Desktop\SSVEP/new_data/np_filt_CAR.npy')
```

```python
import mne

fs = 256
ch_names = ['P1','P3','P5','P7','P10','O2','O7','O8','stim']
ch_types = (['eeg'] * 8) + ['stim']

info = mne.create_info(ch_names, sfreq=fs, ch_types=ch_types)
raw = mne.io.RawArray(np_raw, info) # index 1-10 only for EEG and stim channels
```

```python
raw.plot(duration=50, color={'eeg':'blue'})
```

```python
raw.save("D:\\filt_CAR.fif", overwrite=True)
```
