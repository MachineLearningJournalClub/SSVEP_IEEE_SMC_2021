### Plotting Rocket 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

## Subject 1

```python
subj_1_avg_scores = np.load("subj1_avg_scores.npy")
subj_1_std_scores = np.load("subj1_std_scores.npy")

class_error_1 = 1 - subj_1_avg_scores
diff_std = class_error_1-subj_1_std_scores
```

```python
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rc('legend',fontsize = 20)

fig, ax = plt.subplots(figsize = (15,8))

fig.canvas.draw()

ax.plot(class_error_1)
ax.fill_between(np.array([i for i in range(0,95)]), 
                 diff_std.clip(min = 0), 
                 class_error_1+subj_1_std_scores, alpha = 0.35)
ax.grid(alpha = 0.4)
ax.set_title("ROCKET Classification Error - Subject 1", fontsize = 28)
ax.set_ylabel("Classification Error", fontsize = 24)
ax.set_xlabel("Time [s]" , fontsize = 24)
ax.set_xticks([32,64,96])
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = '1'
labels[1] = '2'
labels[2] = '3'

ax.set_xticklabels(labels)

plt.show()
```
